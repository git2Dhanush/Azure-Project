```python
from pyspark.sql.types import *
from pyspark.sql.functions import *

import os
import sys

project_path = os.path.join(os.getcwd(),"..","..")

sys.path.append(project_path)

from utils.transformations import reuse
```

```python
df_user =spark.readStream.format("cloudFiles")\
                        .option("cloudFiles.format","parquet")\
                        .option("cloudFiles.schemaLocation","abfss://silver@storageaccountdhan.dfs.core.windows.net/DimUser/checkpoint")\
                        .load("abfss://bronze@storageaccountdhan.dfs.core.windows.net/DimUser")
```

```python
df_user = df_user.withColumn("user_name",upper(col("user_name")))
#display(df_user, checkpointLocation = "abfss://silver@storageaccountdhan.dfs.core.windows.net/DimUser/checkpoint")
```

```python
user_obj = reuse()
df_user = user_obj.dropColumns(df_user,['_rescued_data'])
df_user = df_user.dropDuplicates(['user_id'])
#display(df_user, checkpointLocation = "abfss://silver@storageaccountdhan.dfs.core.windows.net/DimUser/checkpoint/offsets")
```

```python
df_user.writeStream.format("delta")\
                    .outputMode("append")\
                    .option("checkpointLocation","abfss://silver@storageaccountdhan.dfs.core.windows.net/DimUser/checkpoint")\
                    .trigger(once=True)\
                    .option("path","abfss://silver@storageaccountdhan.dfs.core.windows.net/DimUser/data")\
                    .toTable("azure_spotify_catalog.silver.DimUser")
                    
```
